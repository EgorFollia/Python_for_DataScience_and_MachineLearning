```python
%matplotlib inline 
import seaborn as sns
import numpy as np

data = sns.load_dataset('tips')
data.head(10)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data = data, estimator = np.std)
```

```python
print(data.total_bill[data.sex == 'Male'].mean())
print(data.total_bill[data.sex == 'Female'].mean())

print(data.total_bill[data.sex == 'Male'].std())
print(data.total_bill[data.sex == 'Female'].std())
```

```python
sns.countplot(x = 'sex', data = data)
```

```python
print(data.sex[data.sex == 'Male'].count())
print(data.sex[data.sex == 'Female'].count())
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = data, hue = 'smoker')
```

```python
np.unique(data.day)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data, hue = 'sex')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', split = True)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, jitter = False)
```

```python
sns.stripplot(x = 'day', y = 'total_bill', data = data, jitter = True, hue = 'sex', dodge = True)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = data, hue = 'sex', dodge = True)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = data)
sns.swarmplot(x = 'day', y = 'total_bill', data = data, color = 'black')
```

```python
sns.factorplot(x = 'day', y = 'total_bill', data = data, kind = 'box')
```
