```python
%matplotlib inline
import seaborn as sns

data = sns.load_dataset('tips')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = data)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = data, hue = 'sex', palette = dict(Male = 'red', Female = 'green'))
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = data, hue = 'sex', markers = ['o', 'v'])
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = data, hue = 'sex', markers = ['o', 'v'], 
           scatter_kws = {'s': 100})
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = data, col = 'sex')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = data, col = 'sex', row = 'smoker')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = data, col = 'day', hue = 'sex',
           aspect = .6, size = 8)
```
