```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 1., 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'g')
axes2.plot(y, x, 'r')
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4)
plt.tight_layout()

for ax in axes:
    ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 1)
plt.tight_layout()

axes[0].plot(x, y)
axes[0].set_title('squared')

axes[1].plot(x, y ** 2)
axes[1].set_title('cubed')

axes[2].plot(y, x)
axes[2].set_title('1/squared')

axes[3].plot(y ** 2, x)
axes[3].set_title('1/cubed')
```

```python
fig = plt.figure(figsize = (8, 2), dpi = 100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 4, ncols = 1, figsize = (10, 8))
fig.tight_layout()

axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'r')
axes[2].plot(x, y ** 2, 'g')
axes[3].plot(y ** 2, x, 'y')
```

```python
fig.savefig('test_pic.png', dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, 'b', label = 'x, y')
ax.plot(y, x, 'r', label = 'y, x')
ax.plot(x, y + 10, 'b--', label = 'x, y + 10')
ax.plot(y + 10, x, 'r--', label = 'y + 10, x')
ax.plot(x, x ** 2, 'g', label = 'x squared')
ax.plot(x, x ** 3, 'g--', label = 'x cubed')
ax.set_xlim(0, 30)
ax.set_ylim(0, 30)

ax.legend(loc = (0.8, 0.4))
```
