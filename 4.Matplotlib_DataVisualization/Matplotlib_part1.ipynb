```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#matplotlib inline
```

```python
x = np.linspace(0, 5, 11, endpoint = True)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional method
plt.plot(x, y, 'r--')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Title')
```

```python
plt.subplot(1, 4, 1) #number of rows - cols - plot we referring to
plt.plot(x, y, 'r')

plt.subplot(1, 4, 2)
plt.plot(x, y ** 2, 'b')

plt.subplot(1, 4, 3)
plt.plot(y, x, 'g--')

plt.subplot(1, 4, 4)
plt.plot(y ** 2, x, 'y--')
```

```python
# Object oriented method
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # takes 4 args - left, bottom, width, height

axes.plot(x, y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 1., 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'g')
axes1.set_xlabel('plot1 x')
axes1.set_ylabel('plot1 y')
axes1.set_title('BigPlot')

axes2.plot(y, x, 'r')
axes2.set_xlabel('plot2 x')
axes2.set_ylabel('plot2 y')
axes2.set_title('SmallPlot')
```
