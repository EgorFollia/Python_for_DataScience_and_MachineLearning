```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

x = np.linspace(0, 5, 11)
y = x ** 2

ax.plot(x, y, color = 'cyan', lw = 5, alpha = 0.5, ls = 'steps', 
        marker = 'o', ms = 20, mfc = 'black', mew = 5, mec = 'red')

ax.set_xlim([0, 10])
ax.set_ylim([0, 30])
```
